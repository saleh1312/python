```python
grad_x = cv2.Sobel(final_mask.astype(np.float32), cv2.CV_32FC1, 1, 0, ksize=3, scale=1, delta=0, borderType=cv2.BORDER_DEFAULT)
grad_y = cv2.Sobel(final_mask.astype(np.float32), cv2.CV_32FC1, 0, 1, ksize=3, scale=1, delta=0, borderType=cv2.BORDER_DEFAULT)
angels=cv2.phase(grad_x,grad_y,angleInDegrees=True)
```

```python
copy_final_mask=np.zeros(final_mask.shape)
points=[]
u=0
for y in range(final_mask.shape[0]):
    for x in range(final_mask.shape[1]):
        ang=angels[y,x]
        if ang !=0 and x < xlimit and y < ylimit:
            ang=ang-90
            if u==30:
                x2 =  int(round(x + 15 * np.cos(ang * 3.14 / 180.0)));
                y2 =  int(round(y + 15 * np.sin(ang * 3.14 / 180.0)));
                copy_final_mask = cv2.line(copy_final_mask, (x,y), (x2,y2), (150), 3)
                copy_final_mask = cv2.circle(copy_final_mask, (x2,y2), 5, (255), 3)
                u=0
                points.append([x,y,ang])
            else:
                u+=1
```
